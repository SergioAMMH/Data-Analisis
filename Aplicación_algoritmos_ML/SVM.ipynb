# Support Vector Machine


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.svm import SVC, SVR

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('df_arabica_clean.csv')
```

## Definimos nuestras X, y

```python
X = df.drop(columns=['Total Cup Points']) 
y = df['Total Cup Points']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
svr = SVR()
```

```python
# Realizar codificación one-hot en las variables categóricas
df_encoded = pd.get_dummies(df)

# Mostrar el DataFrame codificado
print(df_encoded)
df_encoded
```

```python
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
```

```python
X = df_encoded.drop(columns=['Total Cup Points']) 
y = df_encoded['Total Cup Points']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
```

```python
print('Score: ', svr.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio (MSE):", mse)

# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(y_test, y_pred)
print("Coeficiente de determinación (R^2):", r2)
```
